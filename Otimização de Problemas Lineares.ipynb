Considere a seguinte equacao:
8x + y^3 - w + z^2 + x = 256

```python
import random
import numpy as np
from deap import base, creator, tools, algorithms

# Definir o tipo de aptidão (minimização)
creator.create("FitnessMin", base.Fitness, weights=(-1.0,))

# Definir a estrutura do indivíduo (neste caso, uma lista de 4 números)
creator.create("Individual", list, fitness=creator.FitnessMin)

toolbox = base.Toolbox()

# Gerador para criar valores iniciais para as variáveis
toolbox.register("attr_float", random.uniform, -10, 10)

# Inicialização do indivíduo com 4 variáveis
toolbox.register("individual", tools.initRepeat, creator.Individual, toolbox.attr_float, n=4)

# Definir a população como uma lista de indivíduos
toolbox.register("population", tools.initRepeat, list, toolbox.individual)

# Função de avaliação (fitness)
def evaluate(individual):
    x, y, w, z = individual
    equation_result = 8*x + y**3 - w + z**2 + x
    return abs(equation_result - 256),  # Objetivo é minimizar a diferença

toolbox.register("evaluate", evaluate)

# Definir operadores de cruzamento e mutação
toolbox.register("mate", tools.cxTwoPoint)
toolbox.register("mutate", tools.mutGaussian, mu=0, sigma=1, indpb=0.2)

# Seleção de indivíduos para a próxima geração
toolbox.register("select", tools.selTournament, tournsize=4)

def main():
    population = toolbox.population(n=500)
    generations = 1000
    offspring_size = 30

    algorithms.eaSimple(population, toolbox, cxpb=0.7, mutpb=0.2, ngen=generations, stats=None, verbose=True)

    best_individual = tools.selBest(population, k=1)[0]
    best_fitness = best_individual.fitness.values[0]
    best_variables = best_individual

    print("Melhor solução encontrada:")
    print("Variáveis:", best_variables)
    print("Valor da função:", best_fitness)
    print(sum(best_variables))

if __name__ == "__main__":
    main()
```

Uma pequena manufatura produz dois modelos, Standard e Luxo, de um certo produto. Cada unidade do
modelo Standard exige 1 hora de lixação e 1 hora de polimento. Cada unidade do modelo luxo exige 1 hora de
lixação e 4 horas de polimento. A fábrica dispõe de 2 lixadoras e 3 polidoras, cada um trabalhando 40 horas
semanais. As margens de lucro são $24 e $34, respectivamente, para cada unidade Standard e Luxo. Não
existem restrições de demanda para ambos os modelos. Elabore um modelo de programação linear que permita
calcular a produção semanal que maximiza a margem total de lucro do fabricante.

Lucro = 24*x1 + 34*x2

Lixacao = x1 + x2 <=40*2(80)

Polimento =  x1 + x2*4 <= 40 *3 (120)

```python

#Criando um algoritmo de otimização
creator.create("FitnessMax", base.Fitness, weights=(1.0,)) #peso - no caso na parte de otimização
creator.create("Individual", list, fitness=creator.FitnessMax) #Função para otimizar



def constraint1(individual):
    return individual[0] + individual[1] - 80 <= 0 # coloca as restrições lixação
def constraint2(individual):
    return individual[0] + 4 * individual[1] - 120 <= 0 #coloca as restricoes de  Polimento




```

```python
toolbox = base.Toolbox()

# Defina os limites para x1 e x2
x1_bound = (0, 80)
x2_bound = (0, 30)  # Observação: definimos um limite superior para x2 com base na segunda restrição

# Crie atributos para x1 e x2
toolbox.register("attr_x1", random.uniform, x1_bound[0], x1_bound[1])
toolbox.register("attr_x2", random.uniform, x2_bound[0], x2_bound[1])

# Crie indivíduos com dois atributos
toolbox.register("individual", tools.initCycle, creator.Individual, (toolbox.attr_x1, toolbox.attr_x2), n=1)
toolbox.register("population", tools.initRepeat, list, toolbox.individual)

# Função de avaliação (maximizar lucro)
def evaluate(individual):
    x1, x2 = individual
    profit = 24 * x1 + 32 * x2
    return profit,

toolbox.register("evaluate", evaluate)

# Registre as funções de restrição
toolbox.decorate("evaluate", tools.DeltaPenalty(constraint1, (-1e6, 0)))
toolbox.decorate("evaluate", tools.DeltaPenalty(constraint2, (-1e6, 0)))
```

```python
def resultados_salvar(individuo):
  return individuo.fitness.values

estatistica = tools.Statistics(resultados_salvar) #recebe os valores dos calculos
estatistica.register('mean', np.mean) #media
estatistica.register('min', np.min) #minimo
estatistica.register('max', np.max) #valor maximo
```

```python
# Passo 4: Configurar a evolução e executar o algoritmo
population_size = 100
crossover_probability = 0.7
mutation_probability = 0.2
number_of_generations = 50

population = toolbox.population(n=population_size)

toolbox.register("select", tools.selTournament, tournsize=3)
toolbox.register("mate", tools.cxBlend, alpha=0.5)
toolbox.register("mutate", tools.mutGaussian, mu=0, sigma=10, indpb=0.2)

# Imprimir informações sobre cada geração
def log_generation(population, generation, num_generations):
    best_individual = tools.selBest(population, k=1)[0]
    best_profit = evaluate(best_individual)[0]
    print(f"Generation {generation}/{num_generations}: Best Profit = {best_profit:.2f}")
    avg_profit = sum(evaluate(ind)[0] for ind in population) / len(population)
    min_profit = min(evaluate(ind)[0] for ind in population)
    print(f"Generation {generation}/{num_generations}: Best Profit = {best_profit:.2f}, Min Profit = {min_profit:.2f}, Avg Profit = {avg_profit:.2f}")

    # Imprimir os valores de x1 e x2 do melhor indivíduo
    x1, x2 = best_individual
    print(f"x1 = {x1:.2f}, x2 = {x2:.2f}")


# Execute o algoritmo genético com registro de geração
logbook = tools.Logbook()
logbook.header = ["Generation", "Best Profit"]
hof = tools.HallOfFame(1)#pega o primeiro do hall da fama

for generation in range(1, number_of_generations + 1): # roda o lopping de cada geração
    population, _ = algorithms.eaMuPlusLambda(population, toolbox, mu=population_size, lambda_=2*population_size,
                                               cxpb=crossover_probability, mutpb=mutation_probability,
                                               ngen=1, stats=None, halloffame=hof, verbose=False)
    best_individual = tools.selBest(population, k=1)[0]
    best_profit = evaluate(best_individual)[0]
    avg_profit = sum(evaluate(ind)[0] for ind in population) / len(population)
    logbook.record(Generation=generation, **{"Best Profit": best_profit}) #imprime o melhor valor de cada geração
    log_generation(population, generation, number_of_generations) #geracao

# Obtenha o melhor indivíduo após a otimização
best_individual = tools.selBest(population, k=1)[0]
best_profit = evaluate(best_individual)[0]
print("Melhor indivíduo:", best_individual)
print("Melhor lucro:", best_profit)
print("HOF",hof)
```

### Exercicio 3:

Um fazendeiro dispõe de 400 ha cultiváveis com milho, trigo ou soja. Cada hectare de milho exige "2.000 para preparação do terreno, 20 homens-dia de trabalho e gera um lucro de 600". Um hectare de trigo envolve
custos de "2.400 para preparação do terreno, 30 homens-dia de trabalho e dá um lucro de 800". Analogamente,
um hectare de soja exige "1.400, 24 homens-dia e dá um lucro de $400. O fazendeiro dispõe de 800.000" para
cobrir os custos de trabalho e 7.200 homens-dia de mão de obra. Elabore um modelo de programação linear de
forma a calcular a alocação de terra para os vários tipos de cultura com o objetivo de maximizar o lucro total.

milho -> 2000

trigo -> 2400

soja -> 1400

milho -> 20

trigo -> 30

soja -> 24

função:
600 *h1 + 800 *h2 + 400 *h3


h1 + h2 + h3 <= 400

2000*h1 + 2400*h2 + 1400*h3 <= 800.000

20*h1 + 30*h2 + 24*h3 <= 7200






Primeira tentativa de codigo!


```python
import random
from deap import base, creator, tools, algorithms

# Define the optimization problem as a maximization problem
creator.create("FitnessMax", base.Fitness, weights=(1.0,))
creator.create("Individual", list, fitness=creator.FitnessMax)

# Create a toolbox for the optimization problem
toolbox = base.Toolbox()
toolbox.register("attr_float", random.uniform, 0.01, 1.0)  # Genes are real values between 0.01 and 1.0
toolbox.register("individual", tools.initRepeat, creator.Individual, toolbox.attr_float, n=3)  # 3 genes
toolbox.register("population", tools.initRepeat, list, toolbox.individual)

# Define the evaluation function
def evaluate(individual):
    x1, x2, x3 = individual
    objective = x1 * 600 + 800 * x2 + 400 * x3

    # Apply constraints
    constraint1 = x1 + x2 + x3 <= 400
    constraint2 = 2000 * x1 + 2400 * x2 + 1400 * x3 <= 800000
    constraint3 = 20 * x1 + 30 * x2 + 24 * x3 <= 7200

    if constraint1 and constraint2 and constraint3:
        return objective,
    else:
        return -float("inf"),  # Return negative infinity for infeasible solutions

toolbox.register("evaluate", evaluate)
toolbox.register("mate", tools.cxBlend, alpha=0.5)
toolbox.register("mutate", tools.mutGaussian, mu=0, sigma=0.2, indpb=0.2)
toolbox.register("select", tools.selTournament, tournsize=3)

# Create the population and run the genetic algorithm
population = toolbox.population(n=10000)
algorithms.eaMuPlusLambda(population, toolbox, mu=50, lambda_=50, cxpb=0., mutpb=0.3, ngen=10000, stats=None, halloffame=None)

# Get the best individual from the final population
best_individual = tools.selBest(population, k=1)[0]
best_x1, best_x2, best_x3 = best_individual
best_objective = best_x1 * 600 + 800 * best_x2 + 400 * best_x3

print("Best Solution:")
print(f"x1 = {best_x1}, x2 = {best_x2}, x3 = {best_x3}")
print(f"Objective Value: {best_objective}")

```
