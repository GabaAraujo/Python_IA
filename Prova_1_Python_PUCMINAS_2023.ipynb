```python

import tensorflow as tf
print(tf.__version__)

import keras
import pandas as pd
import numpy as np

# import warnings
# warnings.filterwarnings('ignore')



# Entra com os dados
X_treino = np.array(
    [
        [370,7], [220,3], [180,4], [400,7], [380,8],
        [410,9], [220,2], [170,3]
    ]
)
Y_treino = np.array( [ #-1 laranja / # 1 limao
         [-1],[1],[1],[-1],[-1],
         [-1],[1],[1] ] )
X_teste = np.array(
    [
        [380,8], [170,3]
    ]
)
Y_teste = np.array( [[1],[-1]] )




# Monta a rede
from keras.models import Sequential
from keras.layers import Dense

modelo = Sequential() # Inicializa a Rede Neural Artificial

modelo.add(Dense(units = 1, input_dim = X_treino.shape[1]))



# Mostra a rede
modelo.summary()



# Treina a rede
modelo.compile(loss='mae', optimizer = 'adam', metrics=['mae'])


resultado = modelo.fit(X_treino, Y_treino, batch_size = 13, epochs = 2000, validation_data=(X_teste, Y_teste))



# Mostra o Teste
Y_predito = modelo.predict(X_teste)

Y_predito





# Teste da Rede
Y_predito = modelo.predict(X_teste) # Faz o arrendodamento
Y_predito = [ 1 if y>=0.0 else -1 for y in Y_predito ] #positivo +1 / negativo -1
print("Valores Preditos:",Y_predito)

total = 0
correto = 0
errado = 0
for i in Y_predito:
  total=total+1
  if(Y_teste[i,0] == Y_predito[i]):
    correto=correto+1
  else:
    errado=errado+1

print("Total " + str(total))
print("Correto " + str(correto))
print("Errado " + str(errado))



# Mostra graficos
import matplotlib.pyplot as plt

plt.plot(resultado.history['loss'])
plt.plot(resultado.history['val_loss'])
plt.title('Hist√≥rico de Treinamento')
plt.ylabel('Fun√ß√£o de custo')
plt.xlabel('√âpocas de treinamento')
plt.legend(['Erro treino', 'Erro teste'])
plt.show()





# Mostra Pesos
for layerNum, layer in enumerate(modelo.layers):
    weights = layer.get_weights()[0]
    biases = layer.get_weights()[1]

    for toNeuronNum, bias in enumerate(biases):
        print(f'{layerNum}B -> L{layerNum+1}N{toNeuronNum}: {bias}')

    for fromNeuronNum, wgt in enumerate(weights):
        for toNeuronNum, wgt2 in enumerate(wgt):
            print(f'L{layerNum}N{fromNeuronNum} \
                  -> L{layerNum+1}N{toNeuronNum} = {wgt2}')






```

```python
from deap import creator,base,tools,algorithms
import numpy as np
import matplotlib.pyplot as plt


toolbox = base.Toolbox() #cria√ß√£o do nosso ambiente evolutivo.

n = 7 # numero de genes do nosso ambiente - 5 vertices

creator.create("FitnessMin", base.Fitness, weights=(-1.0,)) #a biblioteca deap foi desenvolvida para aceitar algoritmos de maximiza√ß√£o, utilizamos o valor negativo para definir que queremos o valor minimo.
#o peso pode ser dividido em mais variaveis - inclusive concorrentes. Como necessitamos o caminho minimo e este e o unico parametro/peso a ser levado em considera√ß√£o - 1.0 = 100%

creator.create("EstrIndividuo",list,fitness = creator.FitnessMin) #criamos o individuo,ele sera introduzido via lista/grafo e sera baseado nos parametros do FitnessMin

toolbox.register("Genes", np.random.permutation, n) #genes, ser√£o uma permuta√ß√£o baseado em nossos 5 vertices
toolbox.register("Individuos", tools.initIterate, creator.EstrIndividuo, toolbox.Genes) #indivuos ser√£o uma lista - initIterate / moldados pelos genes -
toolbox.register("Populacao", tools.initRepeat, list, toolbox.Individuos) # cria a populacao que tem sao compostos pelos seus individuos


pop = toolbox.Populacao(n=10) #cria a popula√ß√£o de 10 individuos


#inicio da cria√ß√£o do algoritmos de muta√ß√£o/cruzamento e batalha
toolbox.register("mate",tools.cxPartialyMatched) #cruzamento entre os individuos da popula√ß√£o
toolbox.register("mutate", tools.mutShuffleIndexes, indpb = 0.1) #muta√ß√£o dos indivudos
toolbox.register("select", tools.selTournament, tournsize = 2) # individuos disputando entre os valores



dist =        [[0,2,5,4,9999,9999,9999],
              [2,0,2,9999,7,9999,9999],
              [5,2,0,1,4,3,9999],
              [4,9999,1,0,9999,4,9999],
              [9999,7,4,9999,0,1,5],
              [9999,9999,3,4,1,0,7],
              [9999,9999,9999,9999,5,7,0]] #matriz/representa√ß√£o do grafo



#funcao de aptdao -> genes que ele tem de percorrer


def aptidao(individuo):
  f = 0
  for i in range(n-1):
    local1 = individuo[i]
    local2 = individuo[i+1]
    distancia = dist[local1][local2] #percorrer o vetor comparando os indivuduos
    f = f + distancia

  return f,

toolbox.register("evaluate",aptidao)

def resultados_salvar(individuo):
  return individuo.fitness.values


estatistica = tools.Statistics(resultados_salvar) #recebe os valores dos calculos
estatistica.register('mean', np.mean) #media
estatistica.register('min', np.min) #minimo
estatistica.register('max', np.max) #valor maximo



hof = tools.HallOfFame(1)#pega o primeiro do hall da fama


result, log = algorithms.eaSimple(pop,toolbox,cxpb=0.8,mutpb=0.1,stats=estatistica,ngen=100,halloffame=hof,verbose = True)

print("Melhor resultado")
print(hof) #o melhor resultado-> no caso estamos comparando o minimo (-1.0)

print("Melhor geracao")
print(result) #resultados da melhor geracao
```

P(bom motorista | acidente) = P(acidente | bom motorista) * P(bom motorista) / P(acidente)
Sabemos as seguintes probabilidades:

P(acidente | bom motorista) = 0,05


P(bom motorista) = 0,25


P(acidente) = 0,15


Podemos calcular a probabilidade de sofrer um acidente como a m√©dia ponderada da probabilidade de sofrer um acidente dado cada classe de qualidade do motorista, ponderada pela probabilidade de cada classe de qualidade do motorista:

P(acidente) = P(acidente | bom motorista) * P(bom motorista) + P(acidente | motorista m√©dio) * P(motorista m√©dio) + P(acidente | motorista ruim) * P(motorista ruim)



P(acidente) = 0,05 * 0,25 + 0,15 * 0,5 + 0,25 * 0,25


P(acidente) = 0,15


Agora podemos calcular a probabilidade de ser um bom motorista dado que voc√™ sofreu um acidente:

P(bom motorista | acidente) = 0,05 * 0,25 / 0,15


P(bom motorista | acidente) = 0,1667


Portanto, a probabilidade de voc√™ ser um bom motorista dado que voc√™ sofreu um acidente √© 0,1667, ou 16,67%.
